Netflix is one of the most popular media and video streaming platforms. They have over 8000 movies or tv shows available on their platform, as of mid-2021, they have over 200M Subscribers globally. This tabular dataset consists of listings of all the movies and tv shows available on Netflix, along with details such as - cast, directors, ratings, release year, duration, etc.

**About Dataset**

Show_ID-Id of the movie/TV show

type-Movie or TV show

title-Title of the movie/TV show

director-Director of the movie/TV show

Cast-actors/actress who have acted in the movie/TV show

country-Country the movie/TV show belongs to

date_added-Aired dated-Released date on netflix

release year-Original movie/TV show release date

rating-Rating of the movie/TV show

duration-Length of the movie

genre-Genre of the movie

Description-Summary of the movie

### Data Cleaning

```python
## Changing datatype of date_added

## Changing datatype of date_added

date_format = "%d-%b-%y"  # Example format: 14-Aug-20

netflix_data["date_added"] = pd.to_datetime(netflix_data["date_added"], format=date_format)
```

```python
### Checking dtype it is changed or not 
netflix_data["date_added"].dtype
```

```python
#check duplicate
netflix_data.duplicated()
```

```python
### Checking null values
netflix_data.isnull().sum().sort_values(ascending=False)
```

```python
## Cheking percentage of null values
for i in netflix_data.columns:
    if netflix_data[i].isnull().any():
        print(i,"---------",netflix_data[i].isnull().sum()*100/netflix_data.shape[0],"%")
```

```python
#fill the missing values with mode (rating and country)

netflix_data['rating'] = netflix_data['rating'].fillna(netflix_data['rating'].mode()[0])
netflix_data['country'] = netflix_data['country'].fillna(netflix_data['country'].mode()[0])
```

```python
# Get the count of shows/movies aired in each month
month_counts = netflix_data["month_name"].value_counts()

# Create a countplot
plt.figure(figsize=(10, 10))
ax = sns.countplot(x="month_name", data=netflix_data, edgecolor="black")

# Set the title and axis labels
ax.set_title("Distribution of Shows/Movies Aired by Month")
ax.set_xlabel("Month")
ax.set_ylabel("Count")

# Add count values on each bar
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', xytext=(0, 5), textcoords='offset points')

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# Get the count of shows/movies released on each day
day_counts = netflix_data["Day"].value_counts()

# Create a countplot
plt.figure(figsize=(10, 10))
ax = sns.countplot(x="Day", data=netflix_data, edgecolor="black", order=day_counts.index)

# Set the title and axis labels
ax.set_title("Distribution of Shows/Movies Released by Day")
ax.set_xlabel("Day")
ax.set_ylabel(" ")

# Add count values on each bar
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', xytext=(0, 5), textcoords='offset points')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python

# Calculate the percentage of each genre
genre_counts = netflix_data["genres1"].value_counts()
total_count = genre_counts.sum()
genre_percentages = genre_counts / total_count * 100

# Filter genres below 10% and combine them into "Others"
threshold = 1.5
top_genres = genre_percentages[genre_percentages >= threshold]
others_count = genre_percentages[genre_percentages < threshold].sum()
top_genres["Others"] = others_count

# Create a pie chart
plt.figure(figsize=(10, 10), dpi=100)
explode = [0.1] * len(top_genres)  # Explode all the slices by the same amount
plt.pie(top_genres, labels=top_genres.index, autopct="%0.1f%%", explode=explode)

# Set the title
plt.title("Distribution of Aired Genres")

# Display the plot
plt.show()
```

Dramas (17.8%) and comedies (13.8%) were mostly aired that means that were mostly liked by the audiences

```python
# Filter the DataFrame for the year 2020
filtered_data_2020 = netflix_data[netflix_data["release_year"] == 2020]

# Calculate the genre counts for the filtered DataFrame
genre_counts_2020 = filtered_data_2020["genres1"].value_counts()

# Sort the genre counts in descending order
sorted_counts_2020 = genre_counts_2020.sort_values(ascending=False)

# Retrieve the top genres
top_genres_2020 = sorted_counts_2020.head(5)

# Retrieve the top genres
top_genres_2020 = sorted_counts_2020.head(5)

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_genres_2020, labels=top_genres_2020.index, autopct="%0.0f%%", explode=[0.1] * len(top_genres_2020))

# Set the title of the plot
plt.title("Top Viewed Genres in 2020")

# Display the plot
plt.show()
```

In 2020, International TV shows (27%) and Dramas (21%) were the most watched genres, whereas crime TV shows, children and family movies  and comedies accounted for 18% and 17%, respectively.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the average duration for each genre
genre_duration = netflix_data.groupby("genres1")["duration"].mean().sort_values(ascending=False)

# Plot the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(y=genre_duration.index, x=genre_duration.values, palette="viridis")

# Set the title and labels
plt.title("Genres by Duration of TV Shows or Movies")
plt.ylabel("Genres")
plt.xlabel("Average Duration")

# Rotate x-axis labels for better visibility
plt.xticks(rotation=90)

# Display the plot
plt.show()
```

```python
# Filter the dataframe to include only the last 10 years
last_10_years = netflix_data[netflix_data['release_year'] >= netflix_data['release_year'].max() - 10]

# Count the number of shows released per year
show_counts = last_10_years['release_year'].value_counts().sort_index()

# Create the plot
plt.figure(figsize=(10, 6))
show_counts.plot(kind='bar', color='green')

# Set the labels and title
plt.xlabel('Year')
plt.ylabel('Number of Shows Released')
plt.title('Number of Shows Released by Year (Last 10 Years)')

# Show the plot
plt.show()
```
